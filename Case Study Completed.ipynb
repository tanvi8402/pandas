**Q6: Perform below joins between cust_data & cust_demo. 
a. Create data set “cust_leftjoin” using Left Join 
b. Create data set “cust_rightjoin” using right Join 
c. Create data set “cust_innterjoin” using inner Join 
d. Create data set “cust_fulljoin” using full Join**

```python
#a: Left Join
cust_leftjoin=pd.merge(cust_demo,cust_data,how='left',on='ID')
cust_leftjoin
```

```python
#b: Right Join
cust_rightjoin=pd.merge(cust_demo,cust_data,how='right',on='ID')
cust_rightjoin
```

```python
#c: Inner Join
cust_innterjoin=pd.merge(cust_demo,cust_data,how='inner',on='ID')
cust_innterjoin
```

```python
#d: Full Join
cust_fulljoin=pd.merge(cust_demo,cust_data,how='outer',on='ID')
cust_fulljoin
```
